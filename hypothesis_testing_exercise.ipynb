#### Create function `evaluate_test` which prints a conclusion of hypothesis test based on p-value and alpha

PARAMS:
- p (float) - p-value from test
- alpha (float) - significance level


```python
def evaluate_test(p,alpha):
    if p < alpha:
        print("hO is rejected")
    else:
        print("hO is not rejected")
evaluate_test(p,alpha)
```

#### Test the hypothesis that mean of salaries in both companies are equal
- use t-test
- H0: salaries are the same
- HA: salaries are not the same


```python
mean_of_company_A = stat.mean(salaries_company_A)
print(mean_of_company_A)
mean_of_company_B = stat.mean(salaries_company_B)
print(mean_of_company_B)
mean_diff = mean_of_company_A-mean_of_company_B
print(mean_diff)

stats.ttest_ind(salaries_company_A,salaries_company_B)
```

```python
bothsal = np.concatenate((salaries_company_A, salaries_company_B))
salperm = np.random.permutation(bothsal)
perma = salperm[:len(salaries_company_A)]
permb = salperm[len(salaries_company_A):]
print(perma.mean()-permb.mean())

permmeans=[]

for _ in range(1000):
    
    sal_perm = np.random.permutation(bothsal)
    perm_company_A = sal_perm[:len(salaries_company_A)]
    perm_company_B = sal_perm[len(salaries_company_A):]
    
    perm_mean = perm_company_A.mean()-perm_company_B.mean()
    permmeans.append(perm_mean)
    
p = np.sum(np.abs(permmeans) >= mean_diff) / len(permmeans)
print(p)
```
